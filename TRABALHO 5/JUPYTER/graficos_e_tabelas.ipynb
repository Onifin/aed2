```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_astar = pd.read_csv("/content/dados_astar.csv")
df_djk_heap = pd.read_csv("/content/dados_dijkstra_heap.csv")
df_djk_1 = pd.read_csv("/content/dados_dijkstra1.csv")
df_djk_2 = pd.read_csv("/content/dados_dijkstra2.csv")
df_djk_3 = pd.read_csv("/content/dados_dijkstra3.csv")
```

```python
distancia_astar_kmeans = df_astar['distance_astar_kmeans']
distancia_djk_heap_kmeans = df_djk_heap['distance_dijkstra_heap_kmeans']
distancia_djk_kmeans = df_djk_1['distance_dijkstra_kmeans']

distancia_astar_kmeans_constrained = df_astar['distance_astar_kmeans_constrained']
distancia_djk_heap_kmeans_constrained = df_djk_heap['distance_dijkstra_heap_kmeans_constrained']
distancia_djk_kmeans_constrained = df_djk_2['distance_dijkstra_kmeans_constrained']

distancia_astar_random = df_astar['distance_astar_random']
distancia_djk_heap_random = df_djk_heap['distance_dijkstra_heap_random']
distancia_djk_random = df_djk_3['distance_dijkstra_random']

tempo_astar_kmeans = df_astar['time_astar_kmeans']
tempo_djk_heap_kmeans = df_djk_heap['time_dijkstra_heap_kmeans']
tempo_djk_kmeans = df_djk_1['time_dijkstra_kmeans']

tempo_astar_kmeans_constrained = df_astar['time_astar_kmeans_constrained']
tempo_djk_heap_kmeans_constrained = df_djk_heap['time_dijkstra_heap_kmeans_constrained']
tempo_djk_kmeans_constrained = df_djk_2['time_dijkstra_kmeans_constrained']

tempo_astar_random = df_astar['time_astar_random']
tempo_djk_heap_random = df_djk_heap['time_dijkstra_heap_random']
tempo_djk_random = df_djk_3['time_dijkstra_random']

emissao_co2_astar_kmeans = df_astar['emission_astar_kmeans']
emissao_co2_djk_heap_kmeans = df_djk_heap['emission_dijkstra_heap_kmeans']
emissao_co2_djk_kmeans = df_djk_1['emission_dijkstra_kmeans']

emissao_co2_astar_kmeans_constrained = df_astar['emission_astar_kmeans_constrained']
emissao_co2_djk_heap_kmeans_constrained = df_djk_heap['emission_dijkstra_heap_kmeans_constrained']
emissao_co2_djk_kmeans_constrained = df_djk_2['emission_dijkstra_kmeans_constrained']

emissao_co2_astar_random = df_astar['emission_astar_random']
emissao_co2_djk_heap_random = df_djk_heap['emission_dijkstra_heap_random']
emissao_co2_djk_random = df_djk_3['emission_dijkstra_random']
```

```python
# Calcular emissão total de CO2
emissao_co2_total_astar_kmeans = sum(emissao_co2_astar_kmeans)
emissao_co2_total_djk_heap_kmeans = sum(emissao_co2_djk_heap_kmeans)
emissao_co2_total_djk_kmeans = sum(emissao_co2_djk_kmeans)

emissao_co2_total_astar_kmeans_constrained = sum(emissao_co2_astar_kmeans_constrained)
emissao_co2_total_djk_heap_kmeans_constrained = sum(emissao_co2_djk_heap_kmeans_constrained)
emissao_co2_total_djk_kmeans_constrained = sum(emissao_co2_djk_kmeans_constrained)

emissao_co2_total_astar_random = sum(emissao_co2_astar_random)
emissao_co2_total_djk_heap_random = sum(emissao_co2_djk_heap_random)
emissao_co2_total_djk_random = sum(emissao_co2_djk_random)

# Dados das emissões de CO2
dados_co2 = {
    'A* + K-means': emissao_co2_total_astar_kmeans,
    'Dijkstra Heap + K-means': emissao_co2_total_djk_heap_kmeans,
    'Dijkstra + K-means': emissao_co2_total_djk_kmeans,
    'A* + K-means Constrained': emissao_co2_total_astar_kmeans_constrained,
    'Dijkstra Heap + K-means Constrained': emissao_co2_total_djk_heap_kmeans_constrained,
    'Dijkstra + K-means Constrained': emissao_co2_total_djk_kmeans_constrained,
    'A* + Random': emissao_co2_total_astar_random,
    'Dijkstra Heap + Random': emissao_co2_total_djk_heap_random,
    'Dijkstra + Random': emissao_co2_total_djk_random
}

plt.figure(figsize=(14, 8))
plt.bar(dados_co2.keys(), dados_co2.values(), color='black')  # Barras pretas
plt.yscale('log')  # Escala logarítmica no eixo Y
plt.title('Comparação de Emissões de CO2 Totais entre Algoritmos (Escala Logarítmica)')
plt.xlabel('Algoritmos')
plt.ylabel('Emissão Total de CO2 (kg) - Escala Log')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
tempo_total_astar_kmeans = sum(tempo_astar_kmeans)
tempo_total_djk_heap_kmeans = sum(tempo_djk_heap_kmeans)
tempo_total_djk_kmeans = sum(tempo_djk_kmeans)

tempo_total_astar_kmeans_constrained = sum(tempo_astar_kmeans_constrained)
tempo_total_djk_heap_kmeans_constrained = sum(tempo_djk_heap_kmeans_constrained)
tempo_total_djk_kmeans_constrained = sum(tempo_djk_kmeans_constrained)

tempo_astar_random = sum(tempo_astar_random)
tempo_djk_heap_random = sum(tempo_djk_heap_random)
tempo_djk_random = sum(tempo_djk_random)

# Dados dos tempos
dados_tempos = {
    'A* + K-means': tempo_total_astar_kmeans,
    'Dijkstra Heap + K-means': tempo_total_djk_heap_kmeans,
    'Dijkstra + K-means': tempo_total_djk_kmeans,
    'A* + K-means Constrained': tempo_total_astar_kmeans_constrained,
    'Dijkstra Heap + K-means Constrained': tempo_total_djk_heap_kmeans_constrained,
    'Dijkstra + K-means Constrained': tempo_total_djk_kmeans_constrained,
    'A* + Random': tempo_astar_random,
    'Dijkstra Heap + Random': tempo_djk_heap_random,
    'Dijkstra + Random': tempo_djk_random
}

plt.figure(figsize=(14, 8))
plt.bar(dados_tempos.keys(), dados_tempos.values())
plt.yscale('log')  # Escala logarítmica no eixo Y
plt.title('Comparação de Tempos Totais entre Algoritmos (Escala Logarítmica)')
plt.xlabel('Algoritmos')
plt.ylabel('Tempo Total (segundos) - Escala Log')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Calcular as distâncias totais
distancia_total_astar_kmeans = sum(distancia_astar_kmeans)
distancia_total_djk_heap_kmeans = sum(distancia_djk_heap_kmeans)
distancia_total_djk_kmeans = sum(distancia_djk_kmeans)

distancia_total_astar_kmeans_constrained = sum(distancia_astar_kmeans_constrained)
distancia_total_djk_heap_kmeans_constrained = sum(distancia_djk_heap_kmeans_constrained)
distancia_total_djk_kmeans_constrained = sum(distancia_djk_kmeans_constrained)

distancia_total_astar_random = sum(distancia_astar_random)
distancia_total_djk_heap_random = sum(distancia_djk_heap_random)
distancia_total_djk_random = sum(distancia_djk_random)


# Dados das distâncias
dados_distancias = {
    'A* + K-means': distancia_total_astar_kmeans,
    'Dijkstra Heap + K-means': distancia_total_djk_heap_kmeans,
    'Dijkstra + K-means': distancia_total_djk_kmeans,
    'A* + K-means Constrained': distancia_total_astar_kmeans_constrained,
    'Dijkstra Heap + K-means Constrained': distancia_total_djk_heap_kmeans_constrained,
    'Dijkstra + K-means Constrained': distancia_total_djk_kmeans_constrained,
    'A* + Random': distancia_total_astar_random,
    'Dijkstra Heap + Random': distancia_total_djk_heap_random,
    'Dijkstra + Random': distancia_total_djk_random
}

plt.figure(figsize=(14, 8))
plt.bar(dados_distancias.keys(), dados_distancias.values(), color='#E0824C')  # Barras pretas
plt.title('Comparação de Distâncias Totais entre Algoritmos')
plt.xlabel('Algoritmos')
plt.ylabel('Distância Total (m)')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
